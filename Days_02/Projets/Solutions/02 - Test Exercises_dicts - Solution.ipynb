### Exercise 3

1. Create dictionary (dict) `{'posts': 0, 'pages': 0}`
2. Update `posts` to be 100
3. Get the value of `posts` and add 10 and store it to `pages`
- HINT: end result should be `{'posts': 100, 'pages': 110}`

```python
d = {'posts': 0, 'pages': 0}
d
```

```python
d['posts'] = 100
d
```

```python
d['pages'] = d['posts'] + 10
d
```

```python
posts = d['posts']
posts += 10
d['pages'] = posts
d
```
